# WeRateDogs Twitter Account Analysis

Using the WeRateDogs Twitter archive as well as a dog breed prediction algorithm, we can make a few interesting observations about the different dog breeds represented!

First, as seen below, the most common dog breed represented (as identified by our dog breed prediction algorithm based on the image(s) included in the tweets) was a Golden Retriever. Labrador Retrievers follow as second most common, but with less than 3/4 of the amount of Golden Retrievers. See below for more details on the most commonly represented breeds!

### Out of our two favorites, who's more popular?

We have some other information from the WeRateDogs account, including the ratings and favorites for each tweet. Since Goldens and Labs were most common, I thought it'd be interesting to see who was getting better reviews. We can treat the ratings as the review from the WeRateDogs account and the favorites as ratings from their followers.

First, let's take a closer look at our ratings. The majority of the ratings are out of 10, so I focused our histogram below on those ratings (this excluded a few random ratings such as 99/90). As you can see, most dogs get between 10 and 14 out of 10. 

### WeRateDogs favorite breed!

Now that we have an idea of what the ratings look like, we can look at the ratings of our two most common breeds. As you can see below, the labrador retrievers seem to be a real favorite of the WeRateDogs Twitter account with an average rating of almost 14 out of 10! In contrast, Goldens only got around 12 out of 10, and the same for the overall average from all breeds.

### The people's favorite breed!

Now we can look at the number of favorites received on average for each breed. You can see below that there is a strong positive correlation between favorites and retweets, so we can use favorites to get a pretty good idea of which tweets were getting the most interaction.



In contrast to the WeRateDogs account, the people seem to overwhelming favor Golden Retrievers. Labradors also received higher ratings than the overall average, but still not as high as Goldens.

## So which dog breed is most favored?

The people have spoken! Golden Retrievers are posted most often (by far!) and receive way more favorites on average than the other breeds! And who can blame them? Look at that face!
